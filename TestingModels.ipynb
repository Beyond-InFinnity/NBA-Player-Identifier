```python
import os
import numpy as np
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from google.colab import drive
```

```python
fine_tuned_model_path = '/content/drive/My Drive/GA/Capstone/capstone/models/celtics_uncropped_model.h5'
model = load_model(fine_tuned_model_path)
```

```python
uncropped_celtic_dir = '/content/drive/My Drive/GA/Capstone/capstone/images/CelticsImages/FilteredUncropped'
```

```python
test_datagen = ImageDataGenerator(rescale=1./255)

# ceating a generator for the uncropped images
test_generator = test_datagen.flow_from_directory(
    uncropped_celtic_dir,
    target_size=(224, 224),
    batch_size=32,
    class_mode=None,  # I don't need labels for prediction
    shuffle=False
)
```

```python
predictions = model.predict(test_generator)
```

```python
predicted_classes = (predictions > 0.5).astype(int)
```

```python
file_paths = test_generator.filepaths
```

```python
for file_path, predicted_class in zip(file_paths, predicted_classes):
    label = 'Celtic' if predicted_class == 1 else 'Non-Celtic'
    print(f"Image: {os.path.basename(file_path)}, Predicted: {label}")
```
